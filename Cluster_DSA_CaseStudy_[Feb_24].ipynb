```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
wine_data = pd.read_csv('/content/Wine_clust.csv')
wine_data.head()
```

```python
wine_data.info()
```

```python
wine_data.isna().sum()
```

**One-hot encoding**

```python
wine_data = pd.get_dummies(wine_data, dtype='int64')
wine_data.head()
```

**Standardize the data**

```python
from sklearn.preprocessing import StandardScaler
scaled_data=StandardScaler()
scaled_data=scaled_data.fit_transform(wine_data)
```

```python
scaled_data = pd.DataFrame(scaled_data,columns=wine_data.columns)
scaled_data
```

# K-Means Cluster

**Elbow Method**

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init = 'k-means++',n_init=20,random_state=42)
  kmeans.fit(scaled_data)
  wcss.append(kmeans.inertia_)
wcss
```

```python
# The elbow bending point indicates the cluster points
plt.plot(range(1, 11), wcss,marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# KMeans clustering
kmeans=KMeans(n_clusters=3,init = 'k-means++',random_state=42)
kmeans.fit(scaled_data)
y_pred=kmeans.predict(scaled_data)
y_pred
```

```python
plt.scatter(scaled_data['Nonflavanoid_Phenols'],scaled_data['Total_Phenols'],c=y_pred)
plt.title('Nonflavanoid_Phenols vs Total_Phenols')
plt.xlabel('Nonflavanoid_Phenols')
plt.ylabel('Total_Phenols')
plt.show()
```

```python
plt.scatter(scaled_data['Flavanoids'],scaled_data['Total_Phenols'],c=y_pred)
plt.title('Flavanoids vs Total_Phenols')
plt.xlabel('Flavanoids')
plt.ylabel('Total_Phenols')
plt.show()
```

**Silhouette Score**

```python
from sklearn.metrics import silhouette_score
sil_ahc=silhouette_score(scaled_data,y_pred)
sil_ahc
```

# Agglomerative Clustering

```python
import scipy.cluster.hierarchy as sch
# a dendrogram to determine  number of clusters

dendrogram = sch.dendrogram(sch.linkage(scaled_data, method='ward'))
plt.xlabel('Data Points')
plt.ylabel('Euclidean Distance')
plt.title('Dendrogram for Clustering')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='ward')
y_hc = hc.fit_predict(scaled_data)
y_hc
```

**Silhouette Score**

```python
from sklearn.metrics import silhouette_score
sil_ahc=silhouette_score(scaled_data,y_hc)
sil_ahc
```
